```python
import os
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt

base_dir = os.getcwd()
reports_dir = os.path.join(base_dir, "..", "reports")

print(f"Base directory: {base_dir}")
print(f"Reports folder: {reports_dir}")


# 1. Weekly Growth Rate Trend

weekly_path = os.path.join(reports_dir, "weekly_growth.csv")
df_weekly = pd.read_csv(weekly_path, delimiter=';')

fig1 = px.line(
    df_weekly,
    x="date",
    y="weekly_growth_rate",
    title="Weekly COVID-19 Growth Rate Over Time",
    markers=True,
    line_shape="spline",
    color_discrete_sequence=["#FF6347"]
)

fig1.update_layout(
    xaxis_title="Date",
    yaxis_title="Weekly Growth Rate (%)",
    template="plotly_white"
)

fig1.write_html(os.path.join(reports_dir, "trend_weekly_growth.html"))
fig1.show()

# 2. Positivity Rate by Country

positivity_path = os.path.join(reports_dir, "positivity_rate.csv")
df_pos = pd.read_csv(positivity_path, delimiter=';')

fig2 = px.bar(
    df_pos.sort_values(by="positivity_rate", ascending=False),
    x="country_region",
    y="positivity_rate",
    color="positivity_rate",
    color_continuous_scale="Viridis",
    title="COVID-19 Positivity Rate by Country"
)

fig2.update_layout(
    xaxis_title="Country",
    yaxis_title="Positivity Rate (%)",
    template="plotly_white"
)

fig2.write_html(os.path.join(reports_dir, "positivity_rate_chart.html"))
fig2.show()


#3. Top 10 Countries by Total Cases

top10_path = os.path.join(reports_dir, "top_10_countries.csv")
df_top10 = pd.read_csv(top10_path, delimiter=';')

fig3 = px.bar(
    df_top10.sort_values(by="confirmed", ascending=False),
    x="country",
    y="confirmed",
    color="confirmed",
    color_continuous_scale="Reds",
    title="Top 10 Countries with the Highest Number of COVID-19 Cases"
)

fig3.update_layout(
    xaxis_title="Country",
    yaxis_title="Confirmed Cases",
    template="plotly_white"
)

fig3.write_html(os.path.join(reports_dir, "top10_countries_bar.html"))
fig3.show()

#4. Interactive World Map (Global Spread Overview)

world_path = os.path.join(reports_dir, "top_10_countries.csv")
df_world = pd.read_csv(world_path, delimiter=';')

fig4 = px.choropleth(
    df_world,
    locations="country",
    locationmode="country names",
    color="confirmed",
    hover_name="country",
    color_continuous_scale="Reds",
    title="Global COVID-19 Cases (Top 10 Countries)"
)

fig4.update_layout(template="plotly_white")
fig4.write_html(os.path.join(reports_dir, "world_map.html"))
fig4.show()

who_path = os.path.join(reports_dir, "who_regions.csv")

df_who = pd.read_csv(who_path, sep=None, engine="python")
df_who.columns = df_who.columns.str.strip().str.lower()

#5. COVID-19 Cases and Deaths by WHO Region

fig5 = px.bar(
    df_who.melt(
        id_vars="who_region",
        value_vars=["total_cases", "total_deaths"],
        var_name="Metric",
        value_name="Count"
    ),
    x="who_region",
    y="Count",
    color="Metric",
    barmode="group",
    title="COVID-19 Cases and Deaths by WHO Region"
)

fig5.update_layout(
    xaxis_title="WHO Region",
    yaxis_title="Count",
    legend_title_text="Metric",
    template="plotly_white"
)

fig5.write_html(os.path.join(reports_dir, "who_region_analysis.html"))
fig5.show()


# Summary Section

print("All reports loaded successfully.")
print("All visualizations generated in /reports folder.")

from IPython.display import Markdown as md

md("""
# Summary

This notebook presents COVID-19 analytical insights:

1. **Weekly Growth Rate**  tracks pandemic acceleration over time.  
2. **Positivity Rate**  identifies countries with the highest infection rates among tested individuals.  
3. **Top 10 Countries**  highlights nations with the largest case numbers.  
4. **Interactive World Map**  provides a global overview of case distribution.
5. **COVID-19 Cases and Deaths by WHO Region**

*This analysis demonstrates integration of SQL-based data processing, Python data visualization (Plotly/Matplotlib), and a structured analytical pipeline suitable for production or research.*
""")
```
